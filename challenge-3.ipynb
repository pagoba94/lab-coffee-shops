# 1. Read your exported json file from the data folder

```python
import json
import pandas as pd
```

```python
with open("data/coffe_shops.json") as anything:
    new = json.load(anything)
    df = pd.DataFrame(new)
df
```

# 2. Plot a map with the coffee shops

```python
import geopandas as gdp
from cartoframes.viz import Map, Layer, popup_element
import folium
from folium import Choropleth, Circle, Marker, Icon, Map
from folium.plugins import HeatMap, MarkerCluster
import pandas as pd
```

```python
lon = 2.158581
lat = 41.390508


mad_map = Map(location = [lat, lon], zoom_start = 15)

for index, row in df.iterrows():
    
    all_together = {"location": [row["lat"], row["lon"]]}
    icon = Icon (
                icon = "coffee",
                prefix = "fa",
                color = "blue",
                icon_color = "white")

    new_marker = Marker(**all_together, icon = icon)
    
    new_marker.add_to(mad_map)

mad_map
```
