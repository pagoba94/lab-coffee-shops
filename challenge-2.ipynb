# 2. Get names and location of the 10 coffee places by passing Ironhack as your location

```python
def get_coordinates (where):

    url_geocode = f"https://geocode.xyz/{where}?json=1"
    
    try:
        res = requests.get(url_geocode).json()
        return res["longt"], res["latt"]
    
    except:
        print(f"Sorry, no matches for {where}")
        time.sleep(3)
        get_coordinates(where)
```

```python
def foursquare_places (venue, location):
    
    url = f"https://api.foursquare.com/v3/places/search?query={venue}&near={location}"

    headers = {
        "accept": "application/json",
        "Authorization": token
    }

    response = requests.get(url, headers=headers).json()
    return response
```
